```python
import urllib.request #handling URL
from bs4 import BeautifulSoup #Handling html files

response = urllib.request.urlopen("https://en.wikipedia.org/wiki/Nikola_Tesla")
htmlFormat = response.read()
htmlFormat
```

```python
soup = BeautifulSoup(htmlFormat, 'html5lib')# from scraping out not required quotes
#for removing the hmtl tags
text = soup.get_text(strip = True)
tokens = [t for t in text.split()]
tokens
```

```python
import requests
import bs4
res = requests.get('https://quotes.toscrape.com/')
res.text
soup = bs4.BeautifulSoup(res.text,'lxml')
soup.select('.author')
```

```python
authors = set()
for name in soup.select('.author'):
  authors.add(name.text)
authors
```

```python
quotes = []
for quote in soup.select('.text'):
  quotes.append(quote.text)

quotes
```

```python
soup = bs4.BeautifulSoup(res.text,'lxml')
soup.select('.tag-item')
for item in soup.select('.tag-item'):
  print(item.text)
```

```python
from PIL import Image
img = Image.open('sampleImage.jpg')
img
print(img.size)
img
```

```python
# computer vision
import cv2
import imutils
img = cv2.imread('sampleImage.jpg') #reading the image
resizeImg = imutils.resize(img, width = 200)#resize the img
cv2.imwrite('resizedImage.jpg',resizeImg)
```

```python
img = Image.open('resizedImage.jpg')
img
print(img.size)
img
```

```python
#smoothening an image
img = cv2.imread("resizedImage.jpg")
gaussianBlurImg = cv2.GaussianBlur(img,(21,21),0)
cv2.imwrite('bluredImage.jpg',gaussionBlurImg)
```
